# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
df = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv')
df
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/01.png?raw=1)

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(df['month_number'],df['total_profit'], linewidth=2)

axes.set_title('Company profit per month')
axes.set_xlabel('Month Number')
axes.set_ylabel('Total Profit')

axes.set_xticks(df['month_number'])
axes.set_yticks([100000, 200000, 300000, 400000, 500000])

plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/02.png?raw=1)

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(df['month_number'],df['total_profit'], 'ro--',linewidth=5,mfc='k', ms=10)
axes.set_title('Company profit per month')
axes.set_xlabel('Month Number')
axes.set_ylabel('Total Profit')
axes.legend(labels=["Profit data of last year"], loc='lower right')

axes.set_xticks(df['month_number'])
axes.set_yticks([100000, 200000, 300000, 400000, 500000])

plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/03.png?raw=1)

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

line1 = axes.plot(df['month_number'], df['facecream'], 'bo-',linewidth=3)
line2 = axes.plot(df['month_number'], df['facewash'], 'yo-',linewidth=3)
line3 = axes.plot(df['month_number'], df['toothpaste'], 'go-',linewidth=3)
line4 = axes.plot(df['month_number'], df['bathingsoap'], 'ro-',linewidth=3)
line5 = axes.plot(df['month_number'], df['shampoo'], 'mo-',linewidth=3)
line6 = axes.plot(df['month_number'], df['moisturizer'], 'co-',linewidth=3)

axes.set_title('Sale Datas')
axes.set_xlabel('Month Number')
axes.set_ylabel('Sale units in number')
axes.legend(labels=['Face cream Sales Data','Face wash Sales Data','Tooth Paste Sales Data',
                    'Bathing Soap Sales Data','Shampoo Sales Data','Moisturizer'], loc='upper left')

axes.set_xticks(df['month_number'])
axes.set_yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])

plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/04.png?raw=1)

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.scatter(x=df['month_number'],y=df['toothpaste'])
axes.grid(linestyle='--')
axes.set_title('Tooth paste Sales data')
axes.set_xlabel('Month Number')
axes.set_ylabel('Number of units Sold')
axes.legend(labels=['Tooth paste Sales data'], loc='upper left')

axes.set_xticks(df['month_number'])

plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/05.png?raw=1)

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.grid(linestyle='--')
width=0.25
axes.bar(df['month_number']-width/2,df['facecream'],width=width)
axes.bar(df['month_number']+width/2,df['facewash'],width=width)

axes.set_title('Facewash and Facecream Sales Data')
axes.set_xlabel('Month Number')
axes.set_ylabel('Sales units in number')
axes.legend(labels=['Facewash Sales Data','Facecream Sales Data'], loc='upper left')

axes.set_xticks(df['month_number'])

plt.show()
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/06.png?raw=1)

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.grid(linestyle='--')
axes.bar(df['month_number'],df['bathingsoap'])
axes.set_title('Bathingsoap Sales Data')
axes.set_xlabel('Month Number')
axes.set_ylabel('Sales units in number')

axes.set_xticks(df['month_number'])

plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/07.png?raw=1)

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.hist(df['total_profit'],bins=[150000, 175000, 200000, 225000, 250000, 300000, 350000])
axes.set_title('Profit Data')
axes.set_xlabel('Profit range in dollar')
axes.set_ylabel('Actual Profit in dollar')

plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/08.png?raw=1)

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

data= [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
      df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
axes.pie(data, labels=['FaceCream','FaceWash','ToothPaste','BathingSoap','Shampoo','Moisturizer'], autopct='%1.1f%%')
axes.set_title('Sale Data')
axes.legend(loc='lower right',bbox_to_anchor=(1.2,0.))

plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/09.png?raw=1)

```python
fig, axes = plt.subplots(2, sharex=True)

axes[0].plot(df['month_number'], df['bathingsoap'], 'ko-', linewidth=3)
axes[0].set_title('Sales Data of Bathingsoap')

axes[1].plot(df['month_number'], df['facewash'], 'ro-', linewidth=3)
axes[1].set_title('Sales Data of Facewash')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(df['month_number'])

plt.show()
```
